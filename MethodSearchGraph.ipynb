```python
import networkx as nx
import matplotlib.pyplot as plt

# Membuat graf kosong
G = nx.Graph()

# Menambahkan simpul (nodes)
nodes = ['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10', 'V11']
G.add_nodes_from(nodes)

# Menambahkan tepi (edges) dengan biaya (cost) sesuai gambar
edges = [
    ('V1', 'V2', 2), ('V1', 'V3', 1), ('V1', 'V4', 1),
    ('V2', 'V3', 6), ('V2', 'V5', 1),
    ('V3', 'V5', 5), ('V3', 'V6', 7), ('V3', 'V4', 7),
    ('V4', 'V6', 1), ('V4', 'V7', 9),
    ('V5', 'V6', 3), ('V5', 'V8', 2),
    ('V6', 'V8', 7), ('V6', 'V9', 6), ('V6', 'V7', 4),
    ('V7', 'V10', 1), ('V7', 'V9', 3),
    ('V8', 'V9', 9), ('V8', 'V11', 9),
    ('V9', 'V11', 2),
    ('V10', 'V11', 4)
]

# Menambahkan tepi dengan atribut biaya
G.add_weighted_edges_from(edges)

# Fungsi heuristik (gunakan jarak langsung sebagai pendekatan)
def heuristic(a, b):
    pos = {
        'V1': (0, 0), 'V2': (1, 0), 'V3': (2, 0),
        'V4': (0, 1), 'V5': (1, 1), 'V6': (2, 1),
        'V7': (0, 2), 'V8': (1, 2), 'V9': (2, 2),
        'V10': (0, 3), 'V11': (1, 3)
    }
    (x1, y1) = pos[a]
    (x2, y2) = pos[b]
    return ((x1 - x2) ** 2 + (y1 - y2) ** 2) ** 0.5

# Mencari jalur menggunakan A* dari V1 ke V11
path = nx.astar_path(G, 'V1', 'V11', heuristic=heuristic, weight='weight')
length = nx.astar_path_length(G, 'V1', 'V11', heuristic=heuristic, weight='weight')

print("Path: ", path)
print("Path length: ", length)

# Visualisasi graf
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color="#f86e00", node_size=700, font_weight='bold')
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.show()
```

```python
import networkx as nx
import pprint as pp

edges = [
    ('V1', 'V2', {'weight': 2}), ('V1', 'V3', {'weight': 1}), ('V1', 'V4', {'weight': 1}),
    ('V2', 'V3', {'weight': 6}), ('V2', 'V5', {'weight': 1}),
    ('V3', 'V5', {'weight': 5}), ('V3', 'V6', {'weight': 7}), ('V3', 'V4', {'weight': 7}),
    ('V4', 'V6', {'weight': 1}), ('V4', 'V7', {'weight': 9}),
    ('V5', 'V6', {'weight': 3}), ('V5', 'V8', {'weight': 2}),
    ('V6', 'V8', {'weight': 7}), ('V6', 'V9', {'weight': 6}), ('V6', 'V7', {'weight': 4}),
    ('V7', 'V10', {'weight': 1}), ('V7', 'V9', {'weight': 3}),
    ('V8', 'V9', {'weight': 9}), ('V8', 'V11', {'weight': 9}),
    ('V9', 'V11', {'weight': 2}),
    ('V10', 'V11', {'weight': 4})
]

edge_labels = {
    ('V1', 'V2'): 2, ('V1', 'V3'): 1, ('V1', 'V4'): 1,
    ('V2', 'V3'): 6, ('V2', 'V5'): 1,
    ('V3', 'V5'): 5, ('V3', 'V6'): 7, ('V3', 'V4'): 7,
    ('V4', 'V6'): 1, ('V4', 'V7'): 9,
    ('V5', 'V6'): 3, ('V5', 'V8'): 2,
    ('V6', 'V8'): 7, ('V6', 'V9'): 6, ('V6', 'V7'): 4,
    ('V7', 'V10'): 1, ('V7', 'V9'): 3,
    ('V8', 'V9'): 9, ('V8', 'V11'): 9,
    ('V9', 'V11'): 2,
    ('V10', 'V11'): 4
}

#Graf
G = nx.Graph()
G.add_edges_from(edges)

#Visualisasi
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# Menggunakan algoritma Floyd-Warshall jalur terpendek
fw = nx.floyd_warshall(G, weight='weight')

results = {a: dict(b) for a, b in fw.items()}
pp.pprint(results)

#Jalur terpendek dari V1 ke V11
shortest_path_length = results['V1']['V11']
print(f"Shortest path length from V1 to V11: {shortest_path_length}")

plt.show()
```
