```python
import time
import matplotlib.pyplot as plt

# Implement Algoritma Berpangkat pake dnc
def power(x, n):
    if n == 0:
        return 1
    elif n % 2 == 0:
        half_power = power(x, n // 2)
        return half_power * half_power
    else:
        half_power = power(x, (n - 1) // 2)
        return half_power * half_power * x

print (power(2,100))

#implement Algoritma DnC dengan tanpa permisalan
def biasa (x,n):
    if n==0:
        return 1
    elif n%2==0:
        return biasa(x,n/2)*biasa(x,n/2)
    else :
        return biasa(x,(n-1)/2)*biasa(x,(n-1)/2)*x
print(biasa(2,100))

# Implement Naive Algorithm
def naive(x, n):
    result = 1
    for i in range(n):
        result = result * x
    return result

print(naive(2,100))

x = 2

def measure_time(algorithm, x, n):
    start_time = time.time()
    result = algorithm(x, n)
    end_time = time.time()
    return end_time - start_time

n_values = range(1, 1000, 50)

power_times = []
naive_times = []
biasa_times = []

# ngukur waktu gabut..
for n in n_values:
    power_time = measure_time(power, x, n)
    naive_time = measure_time(naive, x, n)
    biasa_time = measure_time(biasa,x,n)

    power_times.append(power_time)
    naive_times.append(naive_time)
    biasa_times.append(biasa_time)


# Grafiknya mas
plt.plot(n_values, power_times, label="Divide Conquer")
plt.plot(n_values, naive_times, label="Naive Algorithm")
plt.plot(n_values,biasa_times,label="dnc biasa")
plt.xlabel("n (Exponent)")
plt.ylabel("Execution Time (seconds)")
plt.title("DnC Algorithms")
plt.legend()
plt.show()
```

```python
import time
import matplotlib.pyplot as plt

# implementasi dengan rekursif
def r_fibonacci(n):
    if n <= 1:
        return n
    return r_fibonacci(n - 1) + r_fibonacci(n - 2)

# Implementasi dengan iteratif
def it_fibonacci(n):
    if n <= 1:
        return n

    dp = [0] * (n + 1)

    dp[0] = 0
    dp[1] = 1

    for i in range(2, n + 1):
        dp[i] = dp[i - 1] + dp[i - 2]

    return dp[n]

def measure_time(algorithm, n):
    start_time = time.time()
    result = algorithm(n)
    end_time = time.time()
    return end_time - start_time

n_values = range(1, 10)

recursive_times = []
iterative_times = []

for n in n_values:
    recursive_time = measure_time(r_fibonacci, n)
    iterative_time = measure_time(it_fibonacci, n)

    recursive_times.append(recursive_time)
    iterative_times.append(iterative_time)

plt.plot(n_values, recursive_times, label="Recursive Fibonacci")
plt.plot(n_values, iterative_times, label="Iterative Fibonacci")
plt.xlabel("n (Input size)")
plt.ylabel("Waktu eksekusi (seconds)")
plt.title("Perbandingan algoritma Fibonacci")
plt.legend()
plt.show()
```
